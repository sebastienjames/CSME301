```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

d = pd.DataFrame()
d["Time taken (seconds)"] = [5, 5, 5, 5, 5, 10, 10, 10, 10, 10, 15, 15, 15, 15, 15, 15]
d["Distance traveled (cm)"] = [77, 73, 88, 87, 90, 116, 116, 120, 118, 114, 169, 174, 168, 167, 169, 171]
d["Deviation from straight line (cm)"] = [-2, 1, 1, 2, 0, -10, -6, -8, -3, -3, 7, 10, -1, -6, -24, -12]
print(d)

plot = sns.scatterplot(d, x="Distance traveled (cm)", y="Deviation from straight line (cm)", hue="Time taken (seconds)")
plot.axhline(y=0)
plot.set_title("Distance traveled (x-axis) and deviation from the center line (y-axis)\n")
sns.move_legend(plot, "lower left")
```

```python
a = pd.DataFrame()

a["theta"] = [-88, -89, -89, -90, -90, 91, 91, 91, 92, 91, 182, 181, 181, 180, 182]
a["speed"] = ["angle"] * 15
a["r"] = [5] * 15
# df = pd.melt(a, id_vars=['r'], var_name='r', value_name='theta')
a["theta"] = np.deg2rad(a["theta"])

r = np.linspace(0, 10, num=100)
df = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})

# Convert the dataframe to long-form or "tidy" format
df = pd.melt(df, id_vars=['r'], var_name='speed', value_name='theta')

print(df)
# Set up a grid of axes with a polar projection
g = sns.FacetGrid(a, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False, legend_out=False)

# Draw a scatterplot onto each axes in the grid
g.map(sns.scatterplot, "theta", "r", legend=False)
```
