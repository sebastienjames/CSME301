```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

d = pd.DataFrame()
d["Time taken (seconds)"] = [15.03, 15.15, 15.30, 15.28, 15.10]
d["Distance from wall (cm)"] = [25.5, 23.1, 22, 23, 22.9]
d["Deviation from straight line (cm)"] = [2, 0.5, -4, -1, 0.5]
print(d)

plot = sns.scatterplot(d, x="Distance from wall (cm)", y="Deviation from straight line (cm)", hue="Time taken (seconds)")
plot.axhline(y=0)
plot.axvline(x=0)
plot.set_title("Distance from the wall (x-axis) and deviation from the center line (y-axis)\n")
```

```python
a = pd.DataFrame()

a["Number of occurences"] = [0, 0, 0, 1, 0, 3, 0, 0, 0, 0, 1]
a["Turn error (degrees)"] = range(5, 16)

print(a)
# fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

plot = sns.barplot(x='Turn error (degrees)', y='Number of occurences', 
            data=a)

plot.set_title("Turn error over 5 trials")
```

```python
b = pd.DataFrame()

b["Turn error (degrees)"] = [10, 10, 10, 15, 50]
b["Number of turns"] = [3.98, 4, 4.02, 6, 7]

plot = sns.lineplot(b, x="Number of turns", y = "Turn error (degrees)")
plot = sns.scatterplot(b, x="Number of turns", y = "Turn error (degrees)", color="red")
plot.set_title("Turn error (degrees) by how many turns were made")
```
